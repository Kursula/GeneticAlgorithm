### Genetic Algorithm optimization 
This notebook show a simple implementation of genetic algorithm for box packing optimization. 

The code is divided to following high level classes: 
- Environment contains all the functionality related to the problem that is to be solved or optimized. 
- Gene and Chromosome are the basic building blocks used to store the solution parameters that are optimized by the genetic algorithm. Chromosome consist of Genes. Single Gene typically represents a single object, feature or parameter in the Environment. 
- Population is the pool of chromosomes, where the chromosomes evolve towards optimized solution. Population lives in generations (iterations) where each generation is evaluated according to the Environment constraints (cost). The population evolution is directed towards lower cost. 
- GeneticAlgorithm class is a generic implementation of the genetic optimization algorithm. 

This structure is developed with the idea that all problem specific code should be located in the Environment, Gene and Chromosome classes. 

```python
%load_ext autoreload
%autoreload 2

import numpy as np
from environment import Environment
from population import Population
from genetic_algorithm import GeneticAlgorithm 
```

### Setup environment and create initial chromosome population 

```python
env = Environment()
env.reset()

rect_names = env.get_rectangle_names()
population = Population(population_size=20, gene_names=rect_names)

env.deploy_chromosome(np.random.choice(population.chromosomes))
env.render()
```

### Run the Genetic Algorithm optimization

```python
ga = GeneticAlgorithm()
ga.fit(2000, env, population)

best_chromosome = sorted(population.chromosomes, key=lambda x: x.cost)[0]
env.deploy_chromosome(best_chromosome)
env.render()
```
